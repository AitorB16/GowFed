```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.ticker as ticker
import configparser
```

```python
MODE = 'fl_AM_ACC' #fl
RUN_NAME = ['F1', 'F2', 'F3', 'F4', 'F5', 'F6']
#PRINT_SCR
#BALANCE_DATA
#TRAIN_SIZE
#TEST_SIZE
```

```python
path = {}

stats = {}
tr_loss = {}
val_loss = {}

cl_id = {}
ds_sizes = {}
accuracies = {}
prcsn = {}
rcl = {}
f1 = {}
roc = {}

for run in RUN_NAME:
    root = 'results/'
    path[run] = root + MODE + '/' + run + '/'
    
    stats = pd.read_csv(path[run] + 'stats.txt', sep='\s+')
    tr_loss[run] = pd.read_csv(path[run] + 'tr_loss.txt', sep='\s+', header = None).values[0]
    val_loss[run] = pd.read_csv(path[run] + 'val_loss.txt', sep='\s+', header = None).values[0]
    
    cl_id[run] = stats['CL_ID'].values
    ds_sizes[run] = stats['TRAIN_DS_SIZE'].values
    accuracies[run] = stats['ACC'].values
    prcsn[run] = stats['PREC'].values
    rcl[run] = stats['RCL'].values
    f1[run] = stats['F1'].values
    roc[run] = stats['ROC'].values   
    
```

```python
for run in RUN_NAME:
    
    '''PLOT DS SIZE'''
    plt.bar(x = cl_id[run], height=ds_sizes[run], log = True)
    plt.title('Dataset size in each node')
    plt.xlabel('Node ID')
    plt.ylabel('Number of instances')
    plt.savefig(path[run] + run + '_inst_num.eps', format='eps')
    plt.show()
    
    '''PLOT LOSS'''
    plt.plot(tr_loss[run])
    plt.plot(val_loss[run])
    plt.legend(["Training Loss", "Validation Loss"])
    plt.title('Model loss')
    plt.xlabel("Number of rounds")
    plt.ylabel("Loss")
    plt.savefig(path[run] + run + '_loss.eps', format='eps')
    plt.show()
    
    '''PLOT PR'''
    plt.rcParams["figure.figsize"] = [9.00, 5.50]
    plt.rcParams["figure.autolayout"] = True
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.grid()
    for i in range(0,len(cl_id[run])):
        x = rcl[run][i]
        y = prcsn[run][i]
        plt.plot(x, y, marker="D", markersize=10, markeredgecolor="black", markerfacecolor="coral")
    plt.title('Precision-Recall Space')
    plt.ylabel('Precision')
    plt.xlabel('Recall')
    plt.savefig(path[run] + run + '_pr.eps', format='eps')
    plt.show()
    
    '''PLOT ACCURACIES'''
    plt.bar(x = cl_id[run], height=accuracies[run], log = True)
    plt.title('Accuracies in each node')
    plt.xlabel('Node ID')
    plt.ylabel('Accuracy')
    plt.savefig(path[run] + run + '_acc.eps', format='eps')
    plt.show()
    
```
